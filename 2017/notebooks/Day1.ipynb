```haskell
{-# LANGUAGE OverloadedStrings #-}

import AdventOfCode.Input
import Data.ByteString.UTF8 (toString)

input <- toString . head <$> day 1
```

```haskell
import Data.List

candidates = filter (\x -> length x > 1) $ group input ++ ["44"]

each = map (\s -> (read (head s:[]) :: Int) * (length s - 1)) candidates
sum each
```

```haskell
pairs inp = zip inp ((drop (length inp `div` 2) inp) ++ (take (length inp `div` 2) inp))

sum $ map (\c -> read (c:[]) :: Int) $ map fst $ filter (\(a,b) -> a == b) $ pairs input
```
