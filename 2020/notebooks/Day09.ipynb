```haskell
{-# LANGUAGE OverloadedStrings #-}

import AdventOfCode
```

```haskell
input <- map read <$> dayLinesString 9 :: IO [Int]
```

```haskell
import Data.List

windows = map ((\l -> (init l, last l)) . take 26) $ filter (\l -> length l >= 26) $ tails input

sumToTwoPreceding :: [Int] -> Int -> (Bool, Int)
sumToTwoPreceding values total = (
    any
        (\remainder -> remainder `elem` values && (remainder*2 /= total))
        (map (\v -> total - v) values)
    , total)
```

```haskell
invalid = snd $ head $ filter (\(p, _) -> not p) $ map (uncurry sumToTwoPreceding) windows
invalid
```

```haskell
tailed = tails input

summed = zip tailed (map (scanl1 (+)) (tails input))
```

```haskell
Just (values, sums) = find (\(vs, ss) -> invalid `elem` ss) summed

range = map fst $ zip values (takeWhile (<= invalid) sums)
minimum range + maximum range
```
