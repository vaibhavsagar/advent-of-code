```python
def count_vowels(string):
    counts = [string.count(vowel) for vowel in 'aeiou']
    return sum(counts)
```

```python
def repeated(string):
    pairs = zip(string, string[1:])
    for pair in pairs:
        if pair[0]==pair[1]: return True
    return False
```

```python
def not_contains(string):
    checks = [substring not in string for substring in ['ab', 'cd', 'pq', 'xy']]
    return all(checks)
```
