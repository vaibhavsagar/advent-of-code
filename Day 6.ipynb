```python
import re
instruction = re.compile('(turn off|turn on|toggle) (\d+?),(\d+?) through (\d+?),(\d+?)$')
def turn_on(grid, start, end):
    x0, y0 = start
    x1, y1 = end
    for y in range(y0,y1+1):
        for x in range(x0,x1+1):
            grid[y][x] = 1

def turn_off(grid, start, end):
    x0, y0 = start
    x1, y1 = end
    for y in range(y0,y1+1):
        for x in range(x0,x1+1):
            grid[y][x] = 0

def toggle(grid, start, end):
    x0, y0 = start
    x1, y1 = end
    for y in range(y0,y1+1):
        for x in range(x0,x1+1):
            grid[y][x] = int(not grid[y][x])
```

```python
def execute(grid, line):
    match = instruction.search(line)
    action = match.group(1)
    start = (int(match.group(2)),int(match.group(3)))
    end = (int(match.group(4)),int(match.group(5)))
    actions = {
        'turn on': turn_on,
        'turn off': turn_off,
        'toggle': toggle
    }
    actions[action](grid, start, end)
```
